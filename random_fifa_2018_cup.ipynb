```python
from random import sample, choice

pots = []
pot1 = ['Russia', 'Germany', 'Brazil', 'Portugal',
        'Argentina', 'Belgium', 'Poland', 'France']
pot2 = ['Spain', 'Peru', 'Switzerland', 'England',
        'Colombia', 'Mexico', 'Uruguay', 'Croatia']
pot3 = ['Denmark', 'Iceland', 'Costa Rica',
        'Sweden', 'Tunisia ', 'Egypt', 'Senegal', 'Iran']
pot4 = ['Serbia', 'Nigeria', 'Australia', 'Japan',
        'Morocco', 'Panama', 'South Korea', 'Saudi Arabia']
pots.extend([pot1, pot2, pot3, pot4])

groups = {'Group A': [], 'Group B': [], 'Group C': [], 'Group D': [], 'Group E': [],
          'Group F': [], 'Group G': [], 'Group H': []}


def group_choice(g_dict):
    rand_group = sample(g_dict.keys(), 1)[0]
    if len(g_dict[rand_group]) < 1:
        return rand_group
    else:
        return group_choice(g_dict)


def team_choice(potX, g_dict):
    rand_team = choice(potX)
    if rand_team not in g_dict.items():
        potX.remove(rand_team)
        return rand_team
    else:
        return team_choice(potX, g_dict)


for group in groups.keys():
    next_group = group_choice(groups)
    for pot in pots:
        groups[next_group].append(team_choice(pot, groups))


def print_title(title):
    print('\t\t' + "{:<15}".format(title))


def playoff(group1, group2, g_dict):
    playoff_1_8.append([g_dict[group1][0], g_dict[group2][1]])
    playoff_1_8.append([g_dict[group2][0], g_dict[group1][1]])


def stage_1_8_shuffle(grps):
    even_pairs = [grps[i] for i in range(0, len(grps)) if i % 2 == 0]
    odd_pairs = [grps[i] for i in range(0, len(grps)) if i % 2 != 0]
    return even_pairs + odd_pairs


def match_winner(pair):
    return choice(pair)


def third_place(pair):
    for team in pair:
        if team not in final_match:
            return team


def print_pair(pair):
    print("{:<17} {} ".format(pair[0], pair[1]))


playoff_1_8 = []
groups_listed = list(groups.keys())

for i in range(0, len(groups.keys()) - 1, 2):
    playoff(groups_listed[i], groups_listed[i + 1], groups)

playoff_1_8 = stage_1_8_shuffle(playoff_1_8)

playoff_1_4 = [(match_winner(playoff_1_8[i]), match_winner(
    playoff_1_8[i + 1])) for i in range(0, len(playoff_1_8) - 1, 2)]

playoff_1_2 = [(match_winner(playoff_1_4[i]), match_winner(
    playoff_1_4[i + 1])) for i in range(0, len(playoff_1_4) - 1, 2)]

final_match = (match_winner(playoff_1_2[0]), match_winner(playoff_1_2[1]))

thirdplace_match = [third_place(pair) for pair in playoff_1_2]

winner = match_winner(final_match)

thirdplace = match_winner(thirdplace_match)

print_title('\t'+'GROUP STAGE')

for k, v in groups.items():
    t1, t2, t3, t4 = v
    print("{:<10} {:<12} {:<12} {:<12} {:<12} ".format(k, t1, t2, t3, t4))

print_title('PLAYOFF')
print_title('ROUND 1/8')

for pair in playoff_1_8:
    print_pair(pair)

print_title('ROUND 1/4')

for pair in playoff_1_4:
    print_pair(pair)

print_title('SEMIFINALS')

for pair in playoff_1_2:
    print_pair(pair)

print_title('THIRD PLACE PLAY-OFF' + '\t\t\t' + 'THIRD PLACE WINNER')

print("{:<17} {:<42} {}".format(thirdplace_match[0], thirdplace_match[1], thirdplace))

print_title('FINAL' + '\t\t\t\t\t' + 'WORLD CUP WINNER')

print("{:<17} {:<42} {}".format(final_match[0], final_match[1], winner))
```
