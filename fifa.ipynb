```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
os.getcwd()
df = pd.read_csv("WorldCups.csv")

#scatter plot goals_scored per year

x = df['GoalsScored']
y = df['Year']
plt.scatter(y,x,s=x*3,c='tab:green',alpha=0.7)
plt.ylabel('Number of goals')
plt.xlabel('Year')
plt.title("Goals per Year")
plt.xticks([1930,1940,1950,1960,1970,1980,1990,2000,2010])
plt.yticks([60,80,100,120,140,160,180,200])
plt.grid(True)

```

```python
#most wins
winners = df['Winner'].value_counts().sort_values().plot(kind='barh',figsize=(12,6))
winners.set_xlabel("Number of wins")
winners.set_ylabel("Country")
winners.set_title('Countries with the most victories')
```

```python
#most 2nd places
winners = df['Runners-Up'].value_counts().sort_values().plot(kind='pie',figsize=(12,6))
winners.set_ylabel(None)
winners.set_title('Countries with the most 2nd places')
```

```python
# attendance by year
attendance = df['Attendance']
att = list()
for item in attendance:
    att.append(float(item)*1000) if len(item) < 8 else att.append(float(item.replace('.','')))
avg_att = sum(att)/len(att)       
plt.bar(df['Year'],att,width=2)
plt.title('Attendance by year')
plt.xlabel('Year')
plt.ylabel('Attendance')
plt.text(2025,2000000,'Average attendance '+str(avg_att))
```
